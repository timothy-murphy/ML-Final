```python
import pandas as pd
from datetime import datetime,date, timedelta
import calendar
```

```python
crimes = pd.read_csv('counts_new_crimes.csv')
housing = pd.read_csv('new_housing.csv')
```

```python
crimes.head()
```

```python
housing.head()
```

```python
housing.Neighborhood = housing.Neighborhood.astype(str)
housing.Year = housing.Year.astype(str)
housing.Month = housing.Month.astype(str)
crimes.Neighborhood = crimes.Neighborhood.astype(str)
crimes.Year = crimes.Year.astype(str)
crimes.Month = crimes.Month.astype(str)
```

```python
merged = pd.merge(housing, crimes, how='inner', on=['Year', 'Month', 'Neighborhood'])
```

```python
merged.head()
```

```python
merged
```

```python
merged = merged.drop(['date', 'crime_count'], axis=1)
```

```python
merged.to_csv('merged_data.csv' , index=False, encoding = "ISO-8859-1")
```
